```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("alphabet_stock_data__1_.csv")
df
```

```python
mask1 = df["Date"] > '2020-01-01'
mask2 = df["Date"] <= '2020-04-30'

df = df[mask1 & mask2]
df
```

```python
ax = df.plot(x='Date', y='Adj Close',  marker = 'o', ylabel = 'Adj Close Values')
ax.legend(['Adj Close'])
ax.set_title('Historical stock prices of Alphabet Inc. over the period 2020-01-01 - 2020-04-30')
ax.grid(True)
```

```python
#istogrammi impilati apertura, chiusura, alti e bassi delle azioni tra date

mask1 = df["Date"] > '2020-01-01'
mask2 = df["Date"] <= '2020-04-30'

df = df[mask1 & mask2]
df.count()

ax = df[['Open', 'Close', 'High', 'Low']].plot.hist(stacked=True, bins=21)
ax.set_title('Historical stock prices oscillations of Alphabet Inc. over the period 2020-01-01 - 2020-04-30')
ax.set_xlabel('Stock prices')
ax.grid(True)

```
