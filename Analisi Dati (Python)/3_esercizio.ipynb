```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv")
df
```

```python
mask1 = df["x"] > 5
mask2 = df["y"] > 5
mask3 = df["z"] > 5

df[mask1 & mask2 & mask3]
```

```python
mask1 = df["cut"] == 'Premium'
mask2 = df["cut"] == 'Ideal'

df[mask1 | mask2]
```
